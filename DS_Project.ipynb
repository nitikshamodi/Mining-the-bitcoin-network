```
import igraph as ig
```

```
G = ig.Graph.Read_GraphML("2010_4") 
```

```
G.summary()
```

```
import random
import numpy as np
dendograms = []
for i in range(45):
  node_id = random.sample(range(0,G.vcount()), int(.99*G.vcount()))
  nodes = G.vs(node_id)
  subG = G.induced_subgraph(nodes)
  print(subG.summary())
  dendograms.append(subG.community_fastgreedy(weights = subG.es['weight']))
  print(type(dendograms[-1]))
```

```
clusters = []
for i in dendograms:
  clusters.append(i.as_clustering())
# clusters = dendograms
```

```
def match(c1, c2):
  c1 = set(c1)
  c2 = set(c2)
  return len(c1.intersection(c2))/max(len(c1), len(c2))
```

```
def best_match(c, t):
  m = -1
  mcluster = []
  for i in t:
    v = match(i, c)
    if v>m:
      m = v
      mcluster = i
  return m, mcluster
```

```
base_tree = G.community_fastgreedy(weights = G.es['weight']).as_clustering()
print(list(base_tree))
```

```
d = {}
dc = {}
ite = 0
for c in list(base_tree):
    if ite%100==0:
      print(ite)
    for i in clusters:
      if len(i)>100:
        bmv, _ = best_match(c, i)
        try:
          d[len(c)] += bmv
          dc[len(c)] += 1
        except:
          d[len(c)] = bmv
          dc[len(c)] = 1
    ite+=1
```
